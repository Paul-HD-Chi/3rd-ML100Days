```python
from sklearn import datasets, metrics

# 如果是分類問題，請使用 DecisionTreeClassifier，若為回歸問題，請使用 DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
# 讀取鳶尾花資料集
iris = datasets.load_iris()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=4)

# 建立模型
clf = DecisionTreeClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
print(iris.feature_names)
print("Feature importance: ", clf.feature_importances_)
```

```python
# estabish decision tree classificaion model
clf = DecisionTreeClassifier(criterion = 'entropy')  # default is 'gini'

# train model on training set
clf.fit(x_train, y_train)

# predict test set
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Feature Importance:\n{clf.feature_importances_}")
```

# boston house-price dataset

```python
# load boston house-price dataset, which is a regression problem
boston = datasets.load_boston()

print(f"boston.data.shape: {boston.data.shape}")
print(f"boston.data:\n{boston.data}")
print(f"boston.target\n{boston.target}")
print(f"boston.DESCR:\n{boston.DESCR}")

# split into training / test sets
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.25, random_state=4)

# estabish decision tree regression model
regressor = DecisionTreeRegressor(criterion = 'mse')  # criterion is 'mse'

# train model on training set
regressor.fit(x_train, y_train)

# predict test set
y_pred = regressor.predict(x_test)

acc = metrics.explained_variance_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Feature Importance:\n{regressor.feature_importances_}")
```

```python
# estabish decision tree regression model
regressor = DecisionTreeRegressor(criterion = 'friedman_mse')  # criterion is 'friedman_mse'

# train model on training set
regressor.fit(x_train, y_train)

# predict test set
y_pred = regressor.predict(x_test)

acc = metrics.explained_variance_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Feature Importance:\n{regressor.feature_importances_}")
```

```python
# estabish decision tree regression model
regressor = DecisionTreeRegressor(criterion = 'mae')  # criterion is 'mae'

# train model on training set
regressor.fit(x_train, y_train)

# predict test set
y_pred = regressor.predict(x_test)

acc = metrics.explained_variance_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Feature Importance:\n{regressor.feature_importances_}")
```

"boston house-price problem" is a regression task. DecisionTreeRegressor is used. And there are three different criteria.  

'mse": Accuracy is <font color='blue'>0.7055042200314208</font>  
'friedman_mse': Accuracy is <font color='blue'>0.7359685510923695</font>  
'mae": Accuracy is <font color='blue'>0.7423465349389415</font>  
  
criterion = 'mae' is much better than the other two.  

# wine dataset

```python
# load wine dataset, which is classification problem
wine = datasets.load_wine()

print(f"wine.data.shape: {wine.data.shape}")
print(f"wine.data:\n{wine.data}")
print(f"wine.target:\n{wine.target}")
print(f"wine.target_names: {wine.target_names}")
print(f"wine.DESCR:\n{wine.DESCR}")

# split into training / test sets
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=4)

# estabish decision tree classificaion model
clf = DecisionTreeClassifier(criterion = 'gini')  # criterion is 'gini'

# train model on training set
clf.fit(x_train, y_train)

# predict test set
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Feature Importance:\n{clf.feature_importances_}")
```

```python
# estabish decision tree classificaion model
clf = DecisionTreeClassifier(criterion = 'entropy')  # criterion is 'entropy'

# train model on training set
clf.fit(x_train, y_train)

# predict test set
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Feature Importance:\n{clf.feature_importances_}")
```

"wine dataset problem" is a classification task. DecisionTreeClassifier is used. And there are two different criteria.  
  
'gini': Accuracy is <font color='blue'>0.8666666666666667</font>  
'entropy': Accuracy is <font color='blue'>0.9777777777777777</font>  
  
criterion = <font color='blue'>'entropy'</font> is much better than criterion = <font color='blue'>'gini'</font>.
