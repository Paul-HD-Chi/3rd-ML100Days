```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
# load wine dataset
wine = datasets.load_wine()

print(f"wine.data.shape: {wine.data.shape}")
print(f"wine.data:\n{wine.data}")
print(f"wine.target:\n{wine.target}")
print(f"wine.target_names: {wine.target_names}")
print(f"wine.DESCR:\n{wine.DESCR}")
```

```python
# split into training set / test set
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=4)

# establish logistic regression model
logreg = linear_model.LogisticRegression()

# train the model on training set
logreg.fit(x_train, y_train)

# predict the test set
y_pred = logreg.predict(x_test)

# verify the accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```

```python
# load boston house-price dataset
boston = datasets.load_boston()

print(f"boston.data.shape: {boston.data.shape}")
print(f"boston.data:\n{boston.data}")
print(f"boston.target:\n{boston.target}")
print(f"boston.feature_names: {boston.feature_names}")
print(f"boston.DESCR:\n{boston.DESCR}")
```

```python
# select RM feature (average number of rooms per dwelling) for simplification and visualization purpose
X = boston.data[:, np.newaxis, 5]

# split into training set / test set
x_train, x_test, y_train, y_test = train_test_split(X, boston.target, test_size=0.3, random_state=4)

# establish linear regression model
regr = linear_model.LinearRegression()

# train the model on training set
regr.fit(x_train, y_train)

# predict the test set
y_pred = regr.predict(x_test)

# check coefficients of linear regression model
print('Coefficients: ', regr.coef_)

# identify the distance of predict and real values using MSE
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")

# draw the scatter and regression model of test set
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
# load breast cancer dataset, which is binary classification task
breast_cancer = datasets.load_breast_cancer()

print(f"breast_cancer.data.shape: {breast_cancer.data.shape}")
print(f"breast_cancer.data:\n{breast_cancer.data}")
print(f"breast_cancer.target:\n{breast_cancer.target}")
print(f"breast_cancer.target_names: {breast_cancer.target_names}")
print(f"breast_cancer.DESCR:\n{breast_cancer.DESCR}")
```

```python
# split into training set / test set
x_train, x_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.3, random_state=4)

# establish logistic regression model
logreg = linear_model.LogisticRegression()

# train the model on training set
logreg.fit(x_train, y_train)

# predict the test set
y_pred = logreg.predict(x_test)

# verify the accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```
