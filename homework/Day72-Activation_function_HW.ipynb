# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
#import matplotlib.pylab as plt
import matplotlib.pyplot as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
# ReLU expression
def relu(x):
    y = np.copy(x)
    y[np.less(x, 0)] = 0
    return y

# Generate a sequence of numbers in range of -10 and 10. 
x = np.linspace(-5,5,100)

plt.plot(x, relu(x), 'b')
plt.grid()
plt.title('ReLU Function')

plt.text(0, 4.5, r'$f(x)=max{(0, x)}$', fontsize=15)

plt.show()
```

```python
def drelu(x):
    return np.greater_equal(x, 0)

plt.plot(x, drelu(x), 'r')
plt.grid()
plt.title('Differential of ReLU Function')

plt.text(-5, 0.2, r"$f'(x)=1 if x >= 0 else 0$", fontsize=15)

plt.show() 
```
